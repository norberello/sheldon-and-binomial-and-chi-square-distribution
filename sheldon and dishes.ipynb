## Making a decision based on maths: the Sheldon Cooper case and not doins the dishes

Leonard, Howard, Rajesh, and Sheldon need to decide who washes the dishes every night. They decided it in a random way: the one choosing the longest stick washes the dishes. Sheldon never washes the dishes as the nights pass on, so the others start thinking that he is cheating, although, they cannot explain how. But, how many nights without washing the dishes are necessary to support the hypothesis that he is actually cheating? Understand that even if this is just a funny example in an attempt for you to understand how statistical hypotheis testing are employed, it works exactly the same no matter how serious the problem is: for example, frequency of murders across four different seasons or the number of persons in a survey in favour or against death penalty.

The probability of not washing the dishes in a group of 4 housemates is 3/4 = 0.75 (75%). Let's simulate what could happen by chance in a complete week ("#" symbols are used before comments, which are independent to the code itself). By the laws of binomial probability, the likelihood of not washing dishes in two continuous days is: $${\left ( \frac{3}{4} \right )^{2}=0.56}$$, three nights: $${\left ( \frac{3}{4} \right )^{3}=0.42}$$, 7 days: $${\left ( \frac{3}{4} \right )^{7}=0.13}$$ and so on. Note that `^` symbol is used to calculate the power of a number and `#` symbols to comment on the code:

```r
0.75^2#probability of two nights in a row = 0.56   
0.75^3#probability of three night in a row = 0.42
0.75^7#probability of seven nights in a row = 0.13
```

At which probability, this is, after how many nights, one should start being suspicious about something going on with Sheldon not having to wash the dishes? Is it a week enough (p = 0.13)? Let's calculate the probability for each of the 31 consecutive nights so that we have a better vision of the problem.

```r
nights<-c(1:31)#create 31 nights in the nights object
probability<-0.75^nights#calculate 31 consecutive probabilities and saves it in the probability object
probability#see probabilities from 1 to 31
```

Have into account something very important: we intuitively understand that as the nights pass on the probabiliy of not washing the dishes just by chance becomes lower and lower. However, notice that we are not psychologically prepared to give an exact number even for two consecutive nights. This means that you need to trust maths as your intuition is not accurate at all. Still, you have the power to give the order to R to do that job for you, R wont complain even if it needs to calculate the probability of a million consecutive nights! Wont take it more than a second. Remember this point when someone claims that statistics are unnecessary in psychology or any other discipline because intution guides people instead to solve any problem.

Now, let's plot the probability expressed as bars for an easier interpretation using the commands `barplot()` and `plot()`. It is a very good advice to plot the data to visually understand what you are doing and the question you are trying to answer, and there are many ways and styles to do this in R, here some:

```r
barplot(probability)
```

```r
#tidy up a bit
barplot(probability, names.arg=nights,col="light blue",
       main="Probability of not washing the dishes in consecutive days")#add the
```

```r
plot(nights,probability)
```

```r
plot(nights,probability, type = "b", pch=19,main="Probability of Sheldon not doing the dishes")
```

```r
plot(nights,probability, type = "l", lwd=2,col="red",
     main="Probability of Sheldon not doing the dishes")#lwd thickness, lty = line type
```

So, now with the picture in hand, the question remains: how many nights are needed for Sheldon to be considered "guilty" as charged? This is certainly an arbitrary decision, note also that maths wont solve the problem, but only help you making a decision. Maths are just another clue to answer the question "is he cheating?). Remember that just using maths we cannot be 100% sure, there is always a probability for any observation. This is, whatever decision you make may have a false positive or a false negative. Still, we need to set a probability for what we observe being unlikely. In general, the threshold is traditionally set in 0.05 (alpha level = 0.05), anything that may happen under such a probability or less is considered "statistically significant", or if you prefer highly unlikely or unprobable. So, let's add 0.05 limit as a horizontal line to the figure:

```r
plot(nights,probability, type = "l", col="green",
     lwd=3,main="Probability of Sheldon not doing the dishes")
abline(h=0.05, col="red",lwd=1,lty=3)#add a horizonatal line for alpha level in 0.05
abline(v=11, col="red",lwd=1,lty=3)#add a vertical line for sucha an alpha level
legend(20,0.18,"alpha = 0.05")#add text
```

**Interpretation**: After ten nights the probability of not washing the dishes goes below 0.05, meaning that it is very unlikely that, just by chance, Sheldon is free off doing dishes in the 11th night. This is how statistical hypothesis testing works: we calculate the probability for our observations occurring by chance and support or not the null hypothesis of no effect or no differences, or in this case not cheating. Notice that these are just maths, we do not know reality for certain. What non-statistical decision shall you make with these numbers?

## A final note (the binomial distribution)

Sheldon's problem is a particular case of the binomial distribution: the probability of occuring something or not by chance, this is, a random variable with two possible outcome values: yes or not. To know more about the binomial distribution <https://www.cimt.org.uk/projects/mepres/alevel/stats_ch5.pdf>. Depending on the problem other statistical solutions can be used.

```r
dbinom(1,1,0.75)#probability of not washing the dishes in a given day
dbinom(10,10,0.75)#probability of not washing the dishes in 10 consecutive days: 3/4^10
dbinom(11,11,0.75)#probability of not washing the dishes in 11 consecutive days: 3/4^11
#notice that these are the exact same values than above
#just for curiosity
dbinom(365,365,0.75)#probability of not washing the dishes for a year: 3/4^365
#notice that these are the exact same values than above
```

Once more, remember this **important** take away message. Although statistical thinking gives a powerful clue to be able to make a decision, they do not warrant what the reality is, this is, whatever low probability we may estimate, Sheldon could still be still be innocent.
